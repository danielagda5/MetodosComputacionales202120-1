```python
def Lagrange(x,xi,i,n):
    
    prod = 1.0
    
    for j in range(n+1): #tengo n+1 puntos 
        if i!=j:
            prod *= (x-xi[j])/(xi[i]-xi[j])
            
    return prod
```

```python
def Poly(x,xi,fxi,n):
    
    summ = 0.
    
    for i in range(n+1):
        
        summ += fxi[i]*Lagrange(x,xi,i,n)
        
    return summ
```

```python
#X1 = [5.,10, 15]
#Y1 = [10.,15,25]

X1 = [0,1,2,3]
Y1 = [0,1,2,3]
```

```python
x = np.linspace(0,5,30)
```

```python
px2 = Poly(x,X1,Y1,3)
```
