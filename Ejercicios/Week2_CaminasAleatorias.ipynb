```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from IPython.display import HTML
from tqdm import tqdm
```

```python
# Ahora llenemos muchas particular
def GetWalk(N = 100, NSteps = 1000):
    
    Particulas = []
    
    for p in range(N):
    
        x = np.zeros(NSteps)
        y = np.zeros(NSteps)
        
        for i in range(1,NSteps):
            theta = 2.*np.pi*np.random.rand()
            x[i] = x[i-1] + np.cos(theta)
            y[i] = y[i-1] + np.sin(theta)
        
        Particulas.append([x,y])
        
    return np.array(Particulas)
```

```python
Particulas = GetWalk()
NFrames = len(Particulas)
print(NFrames)
```

```python
# Definamos funcion de animacion
fig1 = plt.figure(figsize=(6,5))
ax = fig1.add_subplot(1,1,1)

def init():
    ax.set_xlabel('x[m]')
    ax.set_ylabel('y[m]')
    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    
# Definamos funcion de actualizacion
def Update(i):
    plot = ax.clear()
    init()
    
    for p in range(len(Particulas)):
        plot = ax.plot(Particulas[p][0][i],Particulas[p][1][i],color='k',marker='o')
        
    return plot   

plt.close()

Animation = anim.FuncAnimation(fig1, Update, frames=NFrames, init_func=init) 
```

```python
HTML(Animation.to_html5_video())
```
