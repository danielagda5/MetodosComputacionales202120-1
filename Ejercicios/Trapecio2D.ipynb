```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
n = 200 
size = 1.0

x = np.linspace(-size,size,n+1)
y = np.linspace(-size,size,n+1)

dx = (x[-1] - x[0])/(n)
dy = (y[-1] - y[0])/(n)

A = dx*dy
R=1

print(dx,dy)
```

```python
# Definimos funcion
def f(x,y,R=1):
    
    z = R**2 - x**2 - y**2
    
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z)
```

```python
f = np.vectorize(f)
```

```python
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d', elev = 45)
ax.plot_surface(X,Y,Z)
```

```python
def GetIntegral(f,x,y,R,A):
    
    integral = 0.
    
    for i in range(len(x)-1):
        for j in range(len(y)-1):
  
            a = f(x[i],y[j],R)
            b = f(x[i],y[j+1],R)
            c = f(x[i+1],y[j],R)
            d = f(x[i+1],y[j+1],R)
            
            integral += A*(a+b+c+d)/4.
            #print(integral)
            
    return integral
```

```python
print(R,A)
I = GetIntegral(f,x,y,R,A)
```
