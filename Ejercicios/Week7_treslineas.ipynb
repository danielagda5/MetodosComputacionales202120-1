```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib          import rc 
from matplotlib          import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,10,30)

y = np.zeros((3,len(x)))

y[0] = 2*x-2
y[1] = -0.5*x + 0.5
y[2] = 4 - x
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
b = np.array([2,1,4])
At = A.T
print(A.shape)
print(b.shape)
```

```python
M = np.dot(At,A)
print(M)
```

```python
bt = np.dot(At,b)
```

```python
xsol = np.linalg.solve(M,bt)
```

```python
print(xsol)
```

```python
for l in range(y.shape[0]):
    plt.plot(x,y[l])
plt.scatter(xsol[0],xsol[1],color='r')
```

```python
# Esta serÃ­a la menor distancia
np.linalg.norm(np.dot(A,xsol)-b)
```

```python
x1 = np.arange(-5.,5.,0.03)
y1 = np.arange(-5.,5.,0.03)
X,Y = np.meshgrid(x1,y1)

d = np.zeros((len(x1),len(y1)))


for i in range(len(x1)):
    for j in range(len(y1)):
        d[i,j] =  np.linalg.norm(np.dot(A,np.array([x1[i],y1[j]]))-b) 
        
```

```python
ii = np.where( d == np.amin(d) )
print(ii)
```

```python
print(x1[ii[0]],y1[ii[1]])
```

```python
fig1 = plt.figure()
ax = fig1.add_subplot(1,1,1, projection = '3d',elev = 36, azim = 50)
ax.plot_surface(X, Y,d,cmap=cm.coolwarm)
ax.set_xlabel(r'$x$', fontsize=10)
ax.set_ylabel(r'$y$', fontsize=10)
ax.set_zlabel(r'$d(x^{*})$', fontsize=10)
#plt.savefig('Rectas.pdf')
```
